```python
import pandas as pd

students=[]

while True:
  print("--------------------")
  print("Student Management System")
  print("1.Add a new student")
  print("2.Display all students")
  print("3.Calculate the average mark")
  print("4.Find the highest mark")
  print("5.Find the lowest mark")
  print("6.Delete student")
  print("7.Update student information")
  print("8.Load student list from CSV file")
  print("9.Save student list to CSV file")
  print("10.Exit")
  print("--------------------")
  choice =(input("Please enter your choice: "))

  if choice=="1":
    def add_student(students_list):
      name=input("Please enter student's name: ")
      while True :
        ID=input("Please enter student's ID: ")
        for student in students_list:
          if student["ID"]==ID:
            print("Student already exists!")
            print("Try again")
            break
        else:
          break
      mark=float(input("Please enter student's mark: "))
      student={
             "Name":name,
             "ID":ID,
             "Mark":mark
             }
      students_list.append(student)
      print("Student added!")
    add_student(students)


  elif choice=="2":
    def display_students(students_list):
      if len(students_list) ==0:
        print("No students found!")
      else:
        df = pd.DataFrame(students_list)
        print(df)
    display_students(students)


  elif choice=="3":
     def calculate_average_mark(students_list):
       if len(students_list)==0:
         print("No students found!")
       else:
         df = pd.DataFrame(students_list)
         print("The average mark is ",end=" ")
         print(df["Mark"].mean())
     calculate_average_mark(students)


  elif choice=="4":
    def highest_Mark(students_list):
      if len(students_list) == 0:
        print("No students found!")
      else:
        df = pd.DataFrame(students_list)
        highestMark = df['Mark'].max()
        print(f"The highest mark is {highestMark}")
    highest_Mark(students)


  elif choice=="5":
    def lowest_Mark(students_list):
      if len(students_list) ==0:
        print("No students found!")
      else:
        df = pd.DataFrame(students_list)
        lowestMark = df['Mark'].min()
        print(f"The lowest mark is {lowestMark}")
    lowest_Mark(students)


  elif choice =="6":
    def delete_student(students_list):
      if len(students_list)==0:
        print("No students found!")
      else:
       deleteID = input("Please enter the ID of student to delete: ")
       for student in students_list:
         if student["ID"] == deleteID:
            students_list.remove(student)
            print(f"Student with ID {deleteID} removed successfully!")
            break
       else:
         print(f"Student with ID {deleteID} does not exist")
    delete_student(students)


  elif choice=="7":
    def update_student(students_list):
      if len(students_list)==0:
        print("No students found!")
      else:
         updateID = input("please enter ID to update:")
         for student in students_list:
           if student["ID"] == updateID:
              student["Name"] = input("Please enter new name :")
              student["Mark"] = float(input("Please enter new mark :"))
              print("Student updated successfully!")
              break
         else:
           print("Student not found!")
    update_student(students)


  elif choice=="8":
    def load_students(students_list):
      try:
          df = pd.read_csv("students.csv")
          students_list.clear()
          students_list.extend(df.to_dict(orient="records"))
          print("Students loaded successfully!")
      except FileNotFoundError:
          print("File not found, you need to save it first!")
    load_students(students)


  elif choice=="9":
     def save_student(students_list):
        if len(students_list) == 0:
          print("No student to save!")
        else:
           df = pd.DataFrame(students_list)
           df.to_csv("students.csv", index=False)
           print("Students save to file successfully")
     save_student(students)


  elif choice=="10":
    print("Exiting the system...")
    break


  else:
    print("Invalid choice! Please select between 1 and 10.")
```
